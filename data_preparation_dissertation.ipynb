```python
def top_20_percent(group):
    # Calculate the 80th percentile like_count threshold
    threshold = np.percentile(group['like_count'],80)
    # Return comments with like_count >= threshold
    return group[group['like_count'] >= threshold]

# Apply the function to each video group
top_20_df = comm_all.groupby('video_id', group_keys=False).apply(top_20_percent)

top_20_df
```

```python
top_20_df_clean=top_20_df[top_20_df['like_count'] != 0]
top_20_df_clean.info()
```

```python
from googletrans import Translator
from langdetect import detect
import time

# Initialize the translator
translator = Translator()

# Define translation function: only translate non-English text
def translate_to_english(text):
    try:
        lang = detect(text)
        if lang == 'en':
            return lang, text
        translated = translator.translate(text, dest='en')
        return lang, translated.text
    except Exception as e:
        print(f"Translation failed: {text[:30]}... | Error: {e}")
        return None, None


output_file = "translated_comments.csv"  # Output CSV file path

# Read the CSV file
df = top_20_df_clean.copy()

# Prepare lists to store detected languages and translated results
langs = []
translations = []

print("Starting translation...")

for idx, text in enumerate(df["comment"]):
    lang, translated = translate_to_english(str(text))
    langs.append(lang)
    translations.append(translated)

    # Print progress every 10 rows
    if idx % 10 == 0:
        print(f"Processed {idx} comments")

    time.sleep(1.2)  # Slow down to avoid IP blocking

print("Translation completed ")

# Add new columns to the DataFrame
df["language"] = langs
df["translated_comment"] = translations

# Save file
df.to_csv(output_file, index=False, encoding="utf-8-sig")
print(f"Translated results saved to {output_file}")
```

```python
# Change your target_column to the column with review data

target_column = 'translated_comment'
```

```python
# drop data with missing values in the 'content' column (target_column)
df = df.dropna(subset=[target_column])
```

```python
# convert the relevant column to lowercase
df[target_column] = df[target_column].str.lower()
```

```python
# remove contraction
!pip install contractions
import contractions
df[target_column] = df[target_column].map(lambda x: contractions.fix(x))


import re

# Remove new line characters ("/n")

df[target_column] = df[target_column].map(lambda x: re.sub('\s+', ' ', x))

# Remove url link
df[target_column] = df[target_column].apply(lambda x: re.sub('http://\S+|https://\S+', '', x))

# Remove email address
df[target_column] = df[target_column].map(lambda x: re.sub('\S*@\S*\s?', '', x))

# Remove overspace
df[target_column] = df[target_column].map(lambda x: re.sub('\s{2,}', " ", x))
```
